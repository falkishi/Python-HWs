Qs1:
    For the following data set, we will use the t test in order to determine the desired statistical results from the provided set which includes less than 30 elements. 

```python
import numpy as np
import matplotlib.pyplot
import scipy.stats as stats 
from scipy.stats import ttest_1samp
from scipy.stats import norm
```

```python
#The input data set
set1 = [3, -3, 3, 15, 15, -16, 14, 21, 30, -24, 32];

#The size of the sample
sz0 = len(set1);
d0 = sz0 - 1;

#The level of confidence 
con_l0 = 0.9;

alph0 = (1 - con_l0) / 2;
```

```python
#Desired measurments

#The mean of the sample
sm0 = np.mean(set1);
print("The mean of the sample:",sm0);

#The standard deviation of the sample
SD0 = np.std(set1, ddof=1);
print("The standard deviation of the sample:",SD0);

#The standard error of the sample
SE0 = SD0 / (sz0 ** (1/2));
print("The stander error of the sample:",SE0);

#The t-value
tv0 = stats.t.ppf(1 - alph0, d0);
print("The t-value of the sample:",tv0);

#The p-value of the sample
pv0 = 2 * stats.t.cdf(-abs(tv0), d0);
print("p-value: ",pv0);

#The intervala of confidence
CR0 = (sm0 - tv0 * SE0, sm0 + tv0 * SE0);
print("The range for the interval of confidence: ",CR0);
```
