We will use the provided data sheets that show the engagment scores of people that did not demonstrate their knowledge of the course material and people who did demonstrate their knowledge of the course material. 

```python
import numpy as np
import matplotlib.pyplot
import scipy.stats as stats 
from scipy.stats import ttest_1samp
from scipy.stats import norm
```

Qs1:

Null hypothesis: The engagement of students that have demonstrated knowledge is equal to 0.75.

Alternative hypothesis: The engagment of students that have demonstrated knowledge is not equal to 0.75.

Based on the fact that the population is larger than 30, we will be using hypothesis tests.


```python
#people who demonstrated
know_file = open('engagement_1.txt');
set1 = know_file.readlines();
set1 = [float(x) for x in set1];

#people who have not demonstrated
didntknow_file = open('engagement_0.txt');
set0 = didntknow_file.readlines();
set0 = [float(x) for x in set0];
```

Qs2:

```python
#Null hypothesis
Null_hp1 = 0.75;

#Sample size
sz1 = len(set1);
print("Sample size of students that demonstrated knowledge: {}\n".format(sz1));

#Sample mean
sm1 = np.mean(set1);
print("Sample mean of students that demonstrated knowledge: {}\n".format(sm1));

#Sample Standard deviation
SD1 = np.std(set1, ddof=1);
#Sample standard error 
SE1 = SD1/(sz1 ** (0.5));
print("Standard error of students that demonstrated knowledge: {}\n".format(SE1));

#Standered score
Z1 = (sm1-Null_hp1) / (SD1 / (sz1 ** (0.5)));
print("Standard score of students that demonstrated knowledge: {}\n".format(Z1));

#P-value of the sample
PV1 = ttest_1samp(set1, Null_hp1)[1];
print("P-value of students that demonstrated knowledge: {}\n".format(PV1));
```

```python
#The value of alpha is equal to 0.1
if PV1 < 0.1: 
    print("From this we get that the 90% interval of confidence is a refused null hypothesis");
    print("Also, we know that the p-value is not significant\n");
else:
    print("From this we get that the 90% interval of confidence is an allowed null hypothesis");
    print("Also, we know that the p-value is significant\n");

#The value of alpha is equal to 0.05
if PV1 < 0.05:
    print("From this we get that the 95% interval of confidence is a refused null hypothesis");
    print("Also, we know that the p-value is not significant\n")
else:
    print("From this we get that the 95% interval of confidence is an allowed null hypothesis");
    print("Also, we know that the p-value is significant\n");
    
#The value of alpha is equal to 0.01
if PV1 < 0.01:
    print("From this we get that the 99% interval of confidence is a refused null hypothesis");
    print("Also, we know that the p-value is not significant\n");
else:
    print("From this we get that the 99% interval of confidence is an allowed null hypothesis");
    print("Also, we know that the p-value is significant\n");
```

Qs3:

```python
#A level of significance at 0.05
alph = 0.05;
#Value of the standard score
SS_a = norm.ppf(alph);
SE_a = (sm1 - Null_hp1)/SS_a;
#prints SS
print("At a level of significance equal to 0.05, \nThe largest Standart Error is:",SE_a);

#prints MSS
min_size = (SD1 / SE_a) ** 2;
print("The minimum sample size equals:",int(min_size));
```
